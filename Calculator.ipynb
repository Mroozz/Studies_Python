```python
from tkinter import *


# Ta funkcja zwraca ramę aplikacji
def iCalc(source, side):
    storeObj = Frame(source, borderwidth=8, bd=1, bg="gray")
    # A tutaj mamy wyświetlenie objektu
    storeObj.pack(side=side, expand=YES, fill=BOTH)
    return storeObj


# Tutaj mamy przedstawiony przycisk
def button(source, side, text, command=None):
    # Zwraca wartość zapakowaną w przycisku
    storeObj = Button(source, text=text, command=command)
    storeObj.pack(side=side, expand=YES, fill=BOTH)
    return storeObj


# Klasa która dziedziczy po "ramie"
class App(Frame):
    def __init__(self):
        Frame.__init__(self)
        self.option_add("*Font", "Times 15")
        self.pack(expand=YES, fill=BOTH)
        self.master.title("Calculator")

        # Wyświetlacz
        display = StringVar()
        entry = Entry(self, relief=RAISED, textvariable=display, justify='right', bd=10, bg='blue')
        entry.pack(side=TOP)

        # Przycisk do całkowitego wyczyszczenia pola
        erase = iCalc(self, TOP)
        clearBut = "C"
        button(erase, LEFT, clearBut, lambda storeObj=display, q=clearBut: storeObj.set(""))

        # Tablica z klawiszami funkcyjnymi
        for numBut in ("789/", "456*", "123-", "0.+"):
            functionNum = iCalc(self, TOP)
            for char in numBut:
                button(functionNum, LEFT, char, lambda storeObj=display, q=char: storeObj.set(storeObj.get() + q))
        equalButton = iCalc(self, TOP)
        
        # Wynik równania
        for iEqual in "=":
            if iEqual == "=":
                btniEqual = button(equalButton, LEFT, iEqual)
                btniEqual.bind("<ButtonRelease-1>", lambda e, s=self, storeObj=display: s.calc(storeObj), '+')
            else:
                btniEqual = button(equalButton, LEFT, iEqual,
                                   lambda storeObj=display, s='%s' % iEqual: storeObj.set(storeObj.get() + s))
    
    def calc(self, display):
        try:
            display.set(eval(display.get()))
    #W przypadku błędu
        except:
            display.set("ERROR")


if __name__ == '__main__':
    App().mainloop()
        
        
```
